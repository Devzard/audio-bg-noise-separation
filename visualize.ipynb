```python
import utilities
from IPython.display import Audio
```

```python
sampleNumber = 'sample4'

source = "./data/" + sampleNumber + "/source.wav"
noise = "./data/" + sampleNumber + "/noise.wav"
overlaid = "./data/" + sampleNumber + "/overlaid.wav"
```

## Source

```python
src_samples, src_sampling_rate = utilities.load_audio(source)
src_audio_len = len(src_samples)/src_sampling_rate
print("Time Domain Representation")
utilities.plot_waveshow_td_representation(src_samples, src_sampling_rate)

src_freq_data, src_mag = utilities.perform_fft(src_samples, src_sampling_rate)
print("Frequency Domain Representation")
utilities.plot_fd_representation(src_freq_data, src_mag)

src_specgram = utilities.spectrogram(src_samples, src_sampling_rate, max_freq=max(src_freq_data))
print("Spectrogram")
utilities.plot_specgram(src_specgram, src_audio_len, src_sampling_rate)
```

## Noise

```python
noise_samples, noise_sampling_rate = utilities.load_audio(noise)
noise_audio_len = len(noise_samples)/noise_sampling_rate
print("Time Domain Representation")
utilities.plot_waveshow_td_representation(noise_samples, noise_sampling_rate)

noise_freq_data, noise_mag = utilities.perform_fft(noise_samples, noise_sampling_rate)
print("Frequency Domain Representation")
utilities.plot_fd_representation(noise_freq_data, noise_mag)

nosie_specgram = utilities.spectrogram(noise_samples, noise_sampling_rate, max_freq=max(noise_freq_data))
print("Spectrogram")
utilities.plot_specgram(nosie_specgram, noise_audio_len, noise_sampling_rate)
```

## Overlaid

```python
overlaid_samples, overlaid_sampling_rate = utilities.load_audio(overlaid)
overlaid_audio_len = len(overlaid_samples)/overlaid_sampling_rate
print("Time Domain Representation")
utilities.plot_waveshow_td_representation(overlaid_samples, overlaid_sampling_rate)

overlaid_freq_data, overlaid_mag = utilities.perform_fft(overlaid_samples, overlaid_sampling_rate)
print("Frequency Domain Representation")
utilities.plot_fd_representation(overlaid_freq_data, overlaid_mag)

overlaid_specgram = utilities.spectrogram(overlaid_samples, overlaid_sampling_rate, max_freq=max(overlaid_freq_data))
print("Spectrogram")
utilities.plot_specgram(overlaid_specgram, overlaid_audio_len, overlaid_sampling_rate)
```
